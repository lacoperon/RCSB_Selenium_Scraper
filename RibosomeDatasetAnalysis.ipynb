Below here, I'm trying to filter and analyze a dataset of RCSB Ribosome Structures that I've produced

```python
from rcsb_api import *
import matplotlib
matplotlib.rcParams['figure.figsize'] = (30.0, 7.0)
matplotlib.rcParams["font.size"] = "24"
%matplotlib inline  
```

```python
pdb_ids = query_for_pdb_ids()
```

```python
pdb_ids
```

```python
ribosome_candidates = get_info_from_ids(pdb_ids)
```

```python
ribosome_candidates
```

```python
(ribosome_candidates['residueCount'].plot.hist(bins=100).set_xlim(0, max(ribosome_candidates['residueCount'])))
```

```python
list(set(list(ribosome_candidates['experimentalTechnique'])))
```

```python
is_xray = (ribosome_candidates['experimentalTechnique'] == "X-RAY DIFFRACTION")
is_em = (ribosome_candidates['experimentalTechnique'] == "ELECTRON MICROSCOPY")
is_right_method = is_xray | is_em
```

```python
less_than_thousand = ribosome_candidates['residueCount'] < 1000
```

```python
print(ribosome_candidates[is_right_method]['residueCount'].plot.hist(bins=100).set_xlim(0, max(ribosome_candidates['residueCount'])))
```
